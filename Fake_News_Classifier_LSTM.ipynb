```python
import tensorflow
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
vsize = 10000
```

#OneHot Encoding

```python
messages = x.copy()
```

```python
messages['title']
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
  print(i)
  review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
ohrep = [one_hot(words, vsize) for words in corpus]
ohrep
```

```python
sentence_length = 25
embedded = pad_sequences(ohrep, maxn len=sentence_length, padding='pre')
embedded
```

```python
features = 40
model = Sequential()
model.add(Embedding(vsize, features, input_length=sentence_length))
model.add(LSTM(200))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
import numpy as np
x_fin = np.array(embedded)
y_fin = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_fin, y_fin, test_size=0.4, random_state=45)
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=15, batch_size=64)
```
