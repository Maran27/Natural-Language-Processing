```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ["I am a good Boy", 
             "Australia Own the Ashes", 
             "India tour on South Africa", 
             "Omicron the Variant of concern", 
             "Un announces Code Red for Humanity", 
             "I am an Artificial Intelligence Developer"]
```

```python
sentences
```

```python
vocabulary_size = 15000
```

One Hot Encoding

```python
rep = [one_hot(words, vocabulary_size) for words in sentences] # will be getting the index from the dictionary
print(rep)
```

Embedding Layer Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sen_len = 10
embedded = pad_sequences(rep, padding='post', maxlen = sen_len)
print(embedded)
```

```python
dim = 25
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size, 25, input_length=sen_len))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded)
```
